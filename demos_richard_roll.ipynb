# Demostración Matemática del Spread Efectivo de Richard Roll

## 1. Introducción y Supuestos del Modelo

El modelo de Richard Roll (1984) estima el spread bid-ask implícito basándose en la autocorrelación serial negativa de los cambios de precios. Esta autocorrelación surge del "rebote" entre precios bid y ask debido a las transacciones.

### Supuestos fundamentales:

1. **Mercado eficiente**: El valor verdadero del activo sigue un paseo aleatorio
2. **Transacciones alternantes**: Las transacciones alternan entre iniciadores de compra y venta
3. **Spread constante**: El spread bid-ask es constante en el tiempo
4. **Sin costos de inventario**: Solo considera costos de procesamiento de órdenes

## 2. Configuración del Modelo

### Definiciones:

- $V_t$ = Valor verdadero (eficiente) del activo en el tiempo $t$
- $P_t$ = Precio observado de la transacción en el tiempo $t$
- $S$ = Spread bid-ask (constante)
- $Q_t$ = Indicador de dirección de la transacción:
  - $Q_t = +1$ si la transacción es iniciada por un comprador (al precio ask)
  - $Q_t = -1$ si la transacción es iniciada por un vendedor (al precio bid)

### Relación fundamental:

El precio observado está relacionado con el valor verdadero por:

$$P_t = V_t + Q_t \cdot \frac{S}{2}$$

**Interpretación:**
- Si $Q_t = +1$: $P_t = V_t + \frac{S}{2}$ (precio ask)
- Si $Q_t = -1$: $P_t = V_t - \frac{S}{2}$ (precio bid)

## 3. Propiedades del Valor Verdadero

El valor verdadero sigue un paseo aleatorio:

$$V_t = V_{t-1} + \epsilon_t$$

donde $\epsilon_t \sim \text{i.i.d}(0, \sigma_\epsilon^2)$

Por lo tanto:
$$\Delta V_t = V_t - V_{t-1} = \epsilon_t$$

## 4. Derivación de los Cambios de Precios Observados

Los cambios de precios observados son:

$$\Delta P_t = P_t - P_{t-1}$$

Sustituyendo la relación fundamental:

$$\Delta P_t = \left(V_t + Q_t \cdot \frac{S}{2}\right) - \left(V_{t-1} + Q_{t-1} \cdot \frac{S}{2}\right)$$

$$\Delta P_t = (V_t - V_{t-1}) + \frac{S}{2}(Q_t - Q_{t-1})$$

$$\Delta P_t = \epsilon_t + \frac{S}{2}(Q_t - Q_{t-1})$$

## 5. Cálculo de la Varianza de los Cambios de Precios

### Varianza de $\Delta P_t$:

$$\text{Var}(\Delta P_t) = \text{Var}(\epsilon_t) + \text{Var}\left(\frac{S}{2}(Q_t - Q_{t-1})\right)$$

Como $\epsilon_t$ y $(Q_t - Q_{t-1})$ son independientes:

$$\text{Var}(\Delta P_t) = \sigma_\epsilon^2 + \frac{S^2}{4}\text{Var}(Q_t - Q_{t-1})$$

### Cálculo de $\text{Var}(Q_t - Q_{t-1})$:

Asumiendo que $Q_t$ son i.i.d con $P(Q_t = +1) = P(Q_t = -1) = \frac{1}{2}$:

- $E[Q_t] = 0$
- $\text{Var}(Q_t) = E[Q_t^2] - (E[Q_t])^2 = 1 - 0 = 1$

Por independencia:
$$\text{Var}(Q_t - Q_{t-1}) = \text{Var}(Q_t) + \text{Var}(Q_{t-1}) = 1 + 1 = 2$$

### Resultado:

$$\text{Var}(\Delta P_t) = \sigma_\epsilon^2 + \frac{S^2}{2}$$

## 6. Cálculo de la Autocovarianza de Primer Orden

### Covarianza entre $\Delta P_t$ y $\Delta P_{t-1}$:

$$\text{Cov}(\Delta P_t, \Delta P_{t-1}) = E[\Delta P_t \cdot \Delta P_{t-1}] - E[\Delta P_t]E[\Delta P_{t-1}]$$

Como $E[\Delta P_t] = 0$:

$$\text{Cov}(\Delta P_t, \Delta P_{t-1}) = E[\Delta P_t \cdot \Delta P_{t-1}]$$

Sustituyendo las expresiones:

$$E[\Delta P_t \cdot \Delta P_{t-1}] = E\left[\left(\epsilon_t + \frac{S}{2}(Q_t - Q_{t-1})\right)\left(\epsilon_{t-1} + \frac{S}{2}(Q_{t-1} - Q_{t-2})\right)\right]$$

Expandiendo:

$$E[\Delta P_t \cdot \Delta P_{t-1}] = E[\epsilon_t \epsilon_{t-1}] + \frac{S}{2}E[\epsilon_t(Q_{t-1} - Q_{t-2})] + \frac{S}{2}E[\epsilon_{t-1}(Q_t - Q_{t-1})] + \frac{S^2}{4}E[(Q_t - Q_{t-1})(Q_{t-1} - Q_{t-2})]$$

### Evaluación de cada término:

1. $E[\epsilon_t \epsilon_{t-1}] = 0$ (por independencia)
2. $E[\epsilon_t(Q_{t-1} - Q_{t-2})] = 0$ (por independencia)
3. $E[\epsilon_{t-1}(Q_t - Q_{t-1})] = 0$ (por independencia)

Para el cuarto término:
$$E[(Q_t - Q_{t-1})(Q_{t-1} - Q_{t-2})] = E[Q_t Q_{t-1} - Q_t Q_{t-2} - Q_{t-1}^2 + Q_{t-1} Q_{t-2}]$$

Por independencia de los $Q_t$:
- $E[Q_t Q_{t-1}] = E[Q_t]E[Q_{t-1}] = 0$
- $E[Q_t Q_{t-2}] = E[Q_t]E[Q_{t-2}] = 0$
- $E[Q_{t-1}^2] = 1$
- $E[Q_{t-1} Q_{t-2}] = E[Q_{t-1}]E[Q_{t-2}] = 0$

Por lo tanto:
$$E[(Q_t - Q_{t-1})(Q_{t-1} - Q_{t-2})] = 0 - 0 - 1 + 0 = -1$$

### Resultado final:

$$\text{Cov}(\Delta P_t, \Delta P_{t-1}) = \frac{S^2}{4} \cdot (-1) = -\frac{S^2}{4}$$

## 7. Derivación de la Fórmula de Roll

De la ecuación anterior tenemos:

$$\text{Cov}(\Delta P_t, \Delta P_{t-1}) = -\frac{S^2}{4}$$

Despejando $S^2$:

$$S^2 = -4 \cdot \text{Cov}(\Delta P_t, \Delta P_{t-1})$$

Tomando la raíz cuadrada positiva:

$$S = 2\sqrt{-\text{Cov}(\Delta P_t, \Delta P_{t-1})}$$

## 8. Fórmula Final del Spread de Roll

### En términos de autocovarianza:

$$\boxed{S = 2\sqrt{-\text{Cov}(\Delta P_t, \Delta P_{t-1})}}$$

### En términos de autocorrelación:

Como $\text{Cov}(\Delta P_t, \Delta P_{t-1}) = \rho_1 \cdot \text{Var}(\Delta P_t)$:

$$S = 2\sqrt{-\rho_1 \cdot \text{Var}(\Delta P_t)}$$

donde $\rho_1$ es la autocorrelación de primer orden de los cambios de precios.

## 9. Implicaciones Económicas

### Condiciones de validez:

1. **Autocovarianza negativa**: $\text{Cov}(\Delta P_t, \Delta P_{t-1}) < 0$
   - Si es positiva o cero, el modelo no es aplicable
   - Indica ausencia del efecto "rebote" bid-ask

2. **Magnitud del spread**: 
   - Cuanto más negativa la autocovarianza, mayor el spread
   - Refleja la intensidad del trading direccional

### Interpretación práctica:

- **Mercados líquidos**: Autocovarianza ligeramente negativa → Spread pequeño
- **Mercados ilíquidos**: Autocovarianza muy negativa → Spread grande
- **Mercados sin microestructura**: Autocovarianza ≈ 0 → Modelo no aplicable

## 10. Limitaciones del Modelo

1. **Supuesto de transacciones alternantes**: En la realidad, pueden haber secuencias de transacciones del mismo tipo
2. **Spread constante**: Los spreads reales varían con la volatilidad, liquidez y otros factores
3. **Sin efectos de inventario**: Ignora los costos de mantener inventario por parte de los market makers
4. **Mercados perfectamente eficientes**: Asume que no hay otros componentes de autocorrelación

## 11. Extensiones y Variantes

### Modelo de Roll generalizado:

Para períodos con múltiples transacciones, la fórmula se puede ajustar considerando la frecuencia de cambios de dirección.

### Corrección por ruido:

En presencia de ruido microestructural adicional:

$$S = 2\sqrt{-\text{Cov}(\Delta P_t, \Delta P_{t-1}) - \sigma_{\text{ruido}}^2}$$

Esta demostración muestra la elegancia matemática del modelo de Roll: a partir de supuestos simples sobre la microestructura del mercado, deriva una relación directa entre la autocorrelación observable de los precios y el spread bid-ask no observable.
